```python
import zipfile

# The name of the ZIP file you want to create
zipped_filename = "demo_hospital_outpatient_data_clean2.zip"

# The path to the CSV file you want to zip
csv_file_path = "/content/Health Data/Demo Hospital Outpatient Data Clean2_NHC.csv"

# Creating the ZIP file and adding the CSV file to it
with zipfile.ZipFile(zipped_filename, "w", zipfile.ZIP_DEFLATED) as zipf:
    zipf.write(csv_file_path, arcname="Demo_Hospital_Outpatient_Data_Clean2_NHC.csv")
```
