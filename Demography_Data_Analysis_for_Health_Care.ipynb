```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_visit = pd.read_csv('/content/demo_hospital_outpatient_data_clean2.zip')
df_visit.head()
```

```python
visit_gender = df_visit['Gender'].value_counts()
```

```python
visit_Diag = df_visit['Diagnosis'].value_counts()
```

```python
visit_gender.plot(kind = 'bar')
```

```python
visit_Diag.plot(kind = 'bar')
```

```python
visit_gender.plot(kind = 'bar',color =['red','blue'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
visit_Diag.plot(kind = 'bar',color = ['red','blue','pink','green','purple','brown','gray','yellow'])
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df_visit['Age'],bins = 20,color = 'purple')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df_visit['Age'],bins =20,color = 'purple',edgecolor = 'w')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
