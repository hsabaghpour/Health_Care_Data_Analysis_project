```python
import pandas as pd
```

```python
df_visit = pd.read_csv('/content/demo_hospital_outpatient_data_clean2.zip')
```

```python
df_visit
```

```python
df_visit['Visit_Date'][0]
```

```python
type(df_visit['Visit_Date'][0])
```

```python
df_visit['Visit_Date'] = pd.to_datetime(df_visit['Visit_Date'])
```

```python
type(df_visit['Visit_Date'][0])
```

```python
df_visit['Visit_Date'][0]
```

```python
df_visit.set_index('Visit_Date', inplace=True)
```

```python
df_visit
```

```python
df_resampled_year = df_visit.resample('Y').count()
```

```python
df_resampled_year
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_resampled_year.index, df_resampled_year['Patient_ID'], marker='o', linestyle='-')
plt.title('Trend Analysis of Patients Over Time (Yearly)')
plt.xlabel('Year')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()
```

```python
df_resampled_month = df_visit.resample('M').count()
```

```python
df_resampled_month
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_resampled_month.index, df_resampled_month['Patient_ID'], marker='o', linestyle='-')
plt.title('Trend Analysis of Patients Over Time (Monthly)')
plt.xlabel('Month')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()
```
